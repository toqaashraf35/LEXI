```python
import cv2
import os

videos_folder = r"C:\Users\DELL\Downloads\EDITED vid"
output_base = r"C:\Users\DELL\Downloads\Extracted_Frames"

def extract_frames(video_path, output_folder, interval=2):
    os.makedirs(output_folder, exist_ok=True)
    cap = cv2.VideoCapture(video_path)
    fps = int(cap.get(cv2.CAP_PROP_FPS))
    count, frame_id = 0, 0

    while True:
        ret, frame = cap.read()
        if not ret:
            break
        if count % (fps * interval) == 0:
            frame_path = os.path.join(output_folder, f"frame_{frame_id}.jpg")
            cv2.imwrite(frame_path, frame)
            frame_id += 1
        count += 1

    cap.release()
    print(f" Extracted {frame_id} frames from {video_path}")



for file in os.listdir(videos_folder):
    if file.lower().endswith(('.mp4', '.mov', '.avi', '.mkv')):
        video_path = os.path.join(videos_folder, file)

        video_name = os.path.splitext(file)[0]
        output_folder = os.path.join(output_base, video_name)

        print(f" Processing: {file}")
        extract_frames(video_path, output_folder, interval=2)

 
```

```python
from fer.fer import FER
import cv2
import os
import pandas as pd
import numpy as np


def analyze_video_frames(video_frames_folder):
    detector = FER(mtcnn=True)
    rows = []

    for frame_file in sorted(os.listdir(video_frames_folder)):
        if frame_file.lower().endswith((".jpg", ".png", ".jpeg")):
            frame_path = os.path.join(video_frames_folder, frame_file)

            img = cv2.imread(frame_path)
            emotions = detector.detect_emotions(img)

            if emotions:
                probs = emotions[0]["emotions"]
                dominant = max(probs, key=probs.get)

                row = {"frame": frame_file, "dominant_emotion": dominant}
                row.update(probs)
                rows.append(row)

    # لو مفيش ولا وش ظهر
    if len(rows) == 0:
        return None

    df = pd.DataFrame(rows)

 
    summary = {}

    # mean لكل emotion
    emotion_cols = ["angry", "disgust", "fear", "happy", "sad", "surprise", "neutral"]
    for col in emotion_cols:
        summary[f"{col}_mean"] = df[col].mean()

    # dominant distribution
    dominant_counts = df["dominant_emotion"].value_counts(normalize=True)

    for emo in emotion_cols:
        summary[f"dominant_{emo}_percent"] = dominant_counts.get(emo, 0)

    return summary


base_folder = r"C:\Users\DELL\Downloads\Extracted_Frames"

final_results = []

for video_folder in os.listdir(base_folder):
    video_path = os.path.join(base_folder, video_folder)

    if os.path.isdir(video_path):
        print(f" Processing {video_folder} ...")

        summary = analyze_video_frames(video_path)

        if summary is not None:
            summary["video_name"] = video_folder
            final_results.append(summary)


final_df = pd.DataFrame(final_results)
output_excel = os.path.join(base_folder, "FER_summary_all_videos.xlsx")
final_df.to_excel(output_excel, index=False)

print("\n Done! Summary saved in:")
print(output_excel)
```

```python
import os
import cv2
import mediapipe as mp
import pandas as pd
import numpy as np

def analyze_body_folder(frame_folder):
    mp_pose = mp.solutions.pose

    # تشغيل MediaPipe بصيغة with علشان يتقفل لوحده
    with mp_pose.Pose(static_image_mode=True) as pose:

        rows = []

        for frame_file in sorted(os.listdir(frame_folder)):
            frame_path = os.path.join(frame_folder, frame_file)
            if not frame_file.lower().endswith((".jpg", ".png")):
                continue

            img = cv2.imread(frame_path)
            if img is None:
                continue

            # ↓↓↓ تصغير الصورة علشان ما تضربيش OutOfMemory
            img = cv2.resize(img, None, fx=0.3, fy=0.3)

            # تشغيل MediaPipe
            result = pose.process(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))

            if result.pose_landmarks:
                lm = result.pose_landmarks.landmark
                row = {}
                for i in range(len(lm)):
                    row[f"lm_{i}_x"] = lm[i].x
                    row[f"lm_{i}_y"] = lm[i].y
                    row[f"lm_{i}_z"] = lm[i].z
                rows.append(row)

            # مسح الصورة من الذاكرة
            del img

    # مفيش landmarks؟ رجّعي None
    if len(rows) == 0:
        return None

    df = pd.DataFrame(rows)
    mean_row = df.mean(numeric_only=True)
    return mean_row



def analyze_all_videos(base_folder):
    """
    base_folder = فولدر Extracted_Frames اللي جواه فولدر لكل فيديو
    """
    final_results = []

    for video_folder in os.listdir(base_folder):
        video_path = os.path.join(base_folder, video_folder)

        if not os.path.isdir(video_path):
            continue

        print(f" Analyzing body posture for video: {video_folder}")

        result = analyze_body_folder(video_path)

        if result is not None:
            result = result.to_dict()
            result["video_name"] = video_folder
            final_results.append(result)
            print(f" Done: {video_folder}")
        else:
            print(f" No body detected in: {video_folder}")

    # حفظهم في Excel واحد
    final_df2 = pd.DataFrame(final_results)

    output_excel2 = os.path.join(base_folder, "Mediapipe_Body_Summary.xlsx")
    final_df2.to_excel(output_excel2, index=False)

    print("\n All videos processed successfully!")
    print(" Output saved at:", output_excel2)



base_folder = r"C:\Users\DELL\Downloads\Extracted_Frames"
analyze_all_videos(base_folder)
```

```python
import numpy as np
import pandas as pd


final_df["head_tilt"] = abs(final_df["lm_0_y"] - final_df["lm_1_y"])

final_df["head_stability_label"] = np.where(
    final_df["head_tilt"] < 0.08,
    "Stable Head",
    "Unstable Head"
)


final_df["shoulder_width"] = abs(final_df["lm_11_x"] - final_df["lm_12_x"])

final_df["posture_openness_label"] = np.where(
    final_df["shoulder_width"] > 0.15,
    "Open Posture",
    "Closed Posture"
)

final_df["body_stability_label"] = np.where(
    abs(final_df["lm_23_z"]) < 0.3,
    "Stable Body",
    "Unstable Body"
)

# =========================
# 🟢 BODY CONFIDENCE (نصي)
# =========================

def body_confidence(row):
    if (
        row["head_stability_label"] == "Stable Head" and
        row["posture_openness_label"] == "Open Posture" and
        row["body_stability_label"] == "Stable Body"
    ):
        return "High Body Confidence"

    if (
        row["posture_openness_label"] == "Closed Posture" and
        row["body_stability_label"] == "Unstable Body"
    ):
        return "Low Body Confidence"

    return "Moderate Body Confidence"


final_df.loc[:, "body_confidence_label"] = final_df.apply(body_confidence, axis=1)

# =========================
# 🟢 EMOTION ANALYSIS (FER)
# =========================

def emotion_state(row):
    if row["happy_mean"] > 0.4:
        return "Positive / Friendly"

    if row["sad_mean"] > 0.25 or row["fear_mean"] > 0.25:
        return "Anxious / Nervous"

    if row["angry_mean"] > 0.25:
        return "Aggressive / Defensive"

    return "Neutral / Controlled"

final_df["emotional_state_label"] = final_df.apply(emotion_state, axis=1)

# =========================
# 🟢 CONFIDENCE FROM EMOTIONS
# =========================

def emotional_confidence(row):
    if (
        row["dominant_happy_percent"] > 0.3 and
        row["dominant_neutral_percent"] > 0.2
    ):
        return "Emotionally Confident"

    if row["dominant_fear_percent"] > 0.25 or row["dominant_sad_percent"] > 0.3:
        return "Emotionally Insecure"

    return "Emotionally Balanced"

final_df["emotional_confidence_label"] = final_df.apply(emotional_confidence, axis=1)

# =========================
# 🟢 OVERALL CONFIDENCE
# =========================

def overall_confidence(row):
    if (
        row["body_confidence_label"] == "High Body Confidence" and
        row["emotional_confidence_label"] == "Emotionally Confident"
    ):
        return "Very Confident Speaker"

    if (
        row["body_confidence_label"] == "Low Body Confidence" and
        row["emotional_confidence_label"] == "Emotionally Insecure"
    ):
        return "Low Confidence Speaker"

    return "Moderately Confident Speaker"

final_df["overall_confidence_label"] = final_df.apply(overall_confidence, axis=1)

# =========================
# 🟢 ENGAGEMENT LEVEL
# =========================

def engagement(row):
    if row["dominant_happy_percent"] > 0.25 and row["open_posture"] if "open_posture" in row else True:
        return "Highly Engaged"

    if row["dominant_neutral_percent"] > 0.5:
        return "Moderately Engaged"

    return "Low Engagement"

final_df["engagement_label"] = final_df.apply(engagement, axis=1)
```

```python
def random_forest_accuracy(df, X_cols, y_col):
    X = df[X_cols]
    y = df[y_col]

    le = LabelEncoder()
    y_encoded = le.fit_transform(y)

    X_train, X_test, y_train, y_test = train_test_split(
        X, y_encoded, test_size=0.2, random_state=42
    )

    model = Pipeline([
        ("scaler", StandardScaler()),
        ("rf", RandomForestClassifier(
            n_estimators=200,
            max_depth=None,
            random_state=42,
            class_weight="balanced"
        ))
    ])

    model.fit(X_train, y_train)

    y_pred = model.predict(X_test)
    acc = accuracy_score(y_test, y_pred)

    return acc
```

```python
accuracies = []

for target, X_cols in targets.items():
    acc = random_forest_accuracy(complete_df, X_cols, target)
    accuracies.append(acc)
overall_accuracy = np.mean(accuracies)
print("Random_Forest Accuracy:", round(overall_accuracy, 3))
```

```python
from sklearn.tree import DecisionTreeClassifier
def decision_tree_accuracy(df, X_cols, y_col):
    X = df[X_cols]
    y = df[y_col]

    # Encode Y (labels نصية → أرقام)
    le = LabelEncoder()
    y_encoded = le.fit_transform(y)

    X_train, X_test, y_train, y_test = train_test_split(
        X,
        y_encoded,
        test_size=0.2,
        random_state=42
    )

    model = Pipeline([
        ("scaler", StandardScaler()),
        ("dt", DecisionTreeClassifier(
            max_depth=6,
            min_samples_split=5,
            random_state=42
        ))
    ])

    model.fit(X_train, y_train)

    y_pred = model.predict(X_test)
    acc = accuracy_score(y_test, y_pred)

    return acc
```

```python
accuracies = []

for target in targets:
    acc = decision_tree_accuracy(complete_df, X_cols, target)
    accuracies.append(acc)

overall_accuracy = np.mean(accuracies)

print(" DecisionTree Accuracy:", round(overall_accuracy, 3))
```

```python
import cv2
import numpy as np
import pandas as pd
from xgboost import XGBClassifier


X_body = [
    "lm_0_y", "lm_1_y",
    "lm_11_x", "lm_12_x",
    "lm_23_z"
]

X_emotion = [
    "happy_mean", "sad_mean", "angry_mean",
    "dominant_happy_percent",
    "dominant_fear_percent",
    "dominant_neutral_percent"
]

X_hybrid = X_body + X_emotion


# ----------------------------------
# 3️⃣ Main Prediction Function
# ----------------------------------

def predict_video(video_path):

    cap = cv2.VideoCapture(video_path)

    body_data = []
    emotion_data = []

    while True:
        ret, frame = cap.read()
        if not ret:
            break

        # -------------------------
        # Dummy body features
        # -------------------------
        body_features = {
            "lm_0_y": np.random.rand(),
            "lm_1_y": np.random.rand(),
            "lm_11_x": np.random.rand(),
            "lm_12_x": np.random.rand(),
            "lm_23_z": np.random.rand()
        }
        body_data.append(body_features)

        # -------------------------
        # Dummy emotion features
        # -------------------------
        emotion_features = {
            "happy_mean": np.random.rand(),
            "sad_mean": np.random.rand(),
            "angry_mean": np.random.rand(),
            "dominant_happy_percent": np.random.rand(),
            "dominant_fear_percent": np.random.rand(),
            "dominant_neutral_percent": np.random.rand()
        }
        emotion_data.append(emotion_features)

    cap.release()

    if len(body_data) == 0 or len(emotion_data) == 0:
        print("Video empty or features missing!")
        return {}

    body_df = pd.DataFrame(body_data)
    emotion_df = pd.DataFrame(emotion_data)

    body_mean = body_df.mean().to_dict()
    emotion_mean = emotion_df.mean().to_dict()

    final_features = {**body_mean, **emotion_mean}
    final_df = pd.DataFrame([final_features])

    # ✅ تأكدي كل الأعمدة موجودة
    for col in X_hybrid:
        if col not in final_df.columns:
            final_df[col] = 0.0

    predictions = {}

    for label, model_data in models.items():
        model = model_data["model"]
        encoder = model_data["encoder"]

        if label in ["head_stability_label",
                     "posture_openness_label",
                     "body_stability_label",
                     "body_confidence_label"]:
            X_input = final_df[X_body]

        elif label in ["emotional_state_label",
                       "emotional_confidence_label"]:
            X_input = final_df[X_emotion]

        else:
            X_input = final_df[X_hybrid]

        pred_encoded = model.predict(X_input)[0]
        pred_label = encoder.inverse_transform([pred_encoded])[0]
        predictions[label] = pred_label

    return predictions
```
