```python
import librosa
import numpy as np

def extract_audio_features(audio_path):
    y, sr = librosa.load(audio_path, sr=None)
    
    # Tempo (speech rate تقريبًا)
    tempo, _ = librosa.beat.beat_track(y=y, sr=sr)
    
    # Energy (Root Mean Square)
    rms = np.mean(librosa.feature.rms(y=y))
    
    # Pitch (frequency)
    pitches, magnitudes = librosa.piptrack(y=y, sr=sr)
    pitch_values = pitches[magnitudes > np.median(magnitudes)]
    pitch_mean = np.mean(pitch_values)
    pitch_std = np.std(pitch_values)
    
    # نعمل normalization بسيطة
    speech_rate_score = min(1.0, tempo / 250)
    energy_score = min(1.0, rms * 50)
    confidence_score = min(1.0, (energy_score + speech_rate_score) / 2)

    return {
        "speech_rate_score": round(float(speech_rate_score), 3),
        "energy_score": round(float(energy_score), 3),
        "confidence_score": round(float(confidence_score), 3),
        "pitch_mean": round(float(pitch_mean), 3),
        "pitch_std": round(float(pitch_std), 3)
    }
```

```python
import os
import librosa
import numpy as np
import pandas as pd

def extract_audio_features(audio_path):
    y, sr = librosa.load(audio_path)

    
    tempo, _ = librosa.beat.beat_track(y=y, sr=sr)
    speech_rate_score = tempo / 200  

    # الطاقة (energy)
    rms = librosa.feature.rms(y=y).flatten()
    energy_score = np.mean(rms)

    # الثقة (confidence)
    confidence_score = np.clip(1 - np.std(rms) / np.mean(rms), 0, 1)

    # درجة الصوت (pitch)
    pitches, magnitudes = librosa.piptrack(y=y, sr=sr)
    pitch_values = pitches[magnitudes > np.median(magnitudes)]
    pitch_mean = np.mean(pitch_values) if len(pitch_values) > 0 else 0
    pitch_std = np.std(pitch_values) if len(pitch_values) > 0 else 0

    return {
        "speech_rate_score": round(float(speech_rate_score), 3),
        "energy_score": round(float(energy_score), 3),
        "confidence_score": round(float(confidence_score), 3),
        "pitch_mean": round(float(pitch_mean), 3),
        "pitch_std": round(float(pitch_std), 3)
    }

# مسار الفولدر اللي فيه الملفات الصوتية
folder_path = r"C:\Users\DELL\Downloads\audios"

# إنشاء DataFrame فاضي
results = []

# المرور على كل الملفات في الفولدر
for filename in os.listdir(folder_path):
    if filename.lower().endswith(('.mp3', '.wav', '.m4a', '.flac')):
        file_path = os.path.join(folder_path, filename)
        print(f"Analyzing {filename}...")
        try:
            features = extract_audio_features(file_path)
            features["file_name"] = filename
            results.append(features)
        except Exception as e:
            print(f"Error processing {filename}: {e}")

# تحويل النتائج إلى DataFrame
df = pd.DataFrame(results)

# ترتيب الأعمدة
df = df[["file_name", "speech_rate_score", "energy_score", "confidence_score", "pitch_mean", "pitch_std"]]

# عرض النتيجة
print("\n✅ Analysis Complete!\n")
print(df)

# حفظها في ملف Excel (اختياري)
output_path = os.path.join(folder_path, "audio_features.xlsx")
df.to_excel(output_path, index=False)
print(f"\n📁 Saved results to: {output_path}")
```

```python
import os
import librosa
import numpy as np
import pandas as pd

# 📂 مسار الفولدر اللي فيه ملفات الصوت
audio_folder = r"C:\Users\DELL\Downloads\audios"

# 🧾 إنشاء قائمة لتخزين البيانات
data = []

# 🔁 نقرأ كل ملف صوت في الفولدر
for file in os.listdir(audio_folder):
    if file.lower().endswith(('.mp3', '.wav', '.m4a', '.flac')):
        file_path = os.path.join(audio_folder, file)
        print(f"🎧 Analyzing {file}...")

        try:
            # قراءة الملف الصوتي
            y, sr = librosa.load(file_path, sr=None)

            # استخراج MFCC features
            mfcc = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=13)

            # نأخذ المتوسط عبر الزمن (عشان يكون تمثيل ثابت)
            mfcc_mean = np.mean(mfcc, axis=1)

            # إضافة البيانات إلى القائمة
            data.append({
                "audio_file": file,
                "embedding": mfcc_mean.tolist()
            })

        except Exception as e:
            print(f"⚠️ Error processing {file}: {e}")

# 🧠 نحول إلى DataFrame
df2 = pd.DataFrame(data)

# 💾 نحفظ النتائج في Excel
output_path = r"C:\Users\DELL\Downloads\audio_embeddings.xlsx"
df2.to_excel(output_path, index=False)


# نعرض أول 5 صفوف
print(df2.head())

```

```python
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score, mean_absolute_error
import pandas as pd
import numpy as np
import ast

# 🧩 تحميل البيانات
X = pd.read_excel(r"C:\Users\DELL\Downloads\audio_embeddings_sorted.xlsx")
y = pd.read_excel(r"C:\Users\DELL\Downloads\audio_features_sorted.xlsx")

# 🧮 تحويل العمود النصي إلى قائمة أرقام
X['embedding'] = X['embedding'].apply(ast.literal_eval)
X_expanded = pd.DataFrame(X['embedding'].tolist())

# 🎯 تحديد الأعمدة المستهدفة (الـ targets)
target_cols = ["speech_rate_score", "energy_score", "confidence_score", "pitch_mean", "pitch_std"]
y = y[target_cols]

# ✂️ تقسيم البيانات إلى تدريب واختبار
X_train, X_test, y_train, y_test = train_test_split(X_expanded, y, test_size=0.2, random_state=42)

# 🌳 إنشاء الموديل
model = RandomForestRegressor(random_state=42)

# 🏋️‍♀️ تدريب الموديل
model.fit(X_train, y_train)

# 🔮 التنبؤ
y_pred = model.predict(X_test)

# 📏 تقييم الأداء على مجموعة الاختبار
r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
accuracy = (1 - (mae / y_test.mean().mean())) * 100

print(" Performance on Test Set:")
print(f" R² Score: {r2:.3f}")
print(f" Mean Absolute Error: {mae:.3f}")
print(f" Approximate Accuracy: {accuracy:.2f}%")


print("\n Cross-Validation Results:")
cv_scores = cross_val_score(model, X_expanded, y, cv=5, scoring='r2')

print("Cross-validation R² scores:", cv_scores)
print("Mean R² (CV):", np.mean(cv_scores))
print("Std Dev R² (CV):", np.std(cv_scores))
```

```python
import os
import librosa
import numpy as np
import pandas as pd
import joblib
import subprocess

def extract_audio_ffmpeg(video_path, audio_path="temp_audio.mp3"):
    command = [
        "ffmpeg",
        "-i", video_path,   # مسار الفيديو
        "-vn",              # تجاهل الفيديو
        "-acodec", "mp3",  # صيغة WAV
        "-ar", "44100",     # sample rate
        "-ac", "1",         # mono
        audio_path
    ]
    subprocess.run(command, stdout=subprocess.DEVNULL, stderr=subprocess.STDOUT)
    return audio_path

def predict_scores_from_video_ffmpeg(video_path):
    model_path = r"C:/Users/DELL/Downloads/Graduation_project/xgboost_model.pkl"

    # استخراج الصوت
    audio_path = extract_audio_ffmpeg(video_path)

    # قراءة الصوت واستخراج MFCC
    y, sr = librosa.load(audio_path, sr=None)
    mfcc = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=13)
    mfcc_mean = np.mean(mfcc, axis=1)

    X = pd.DataFrame([mfcc_mean])
    #X.columns = [f"mfcc_{i+1}" for i in range(X.shape[1])]

    model = joblib.load(model_path)
    y_pred = model.predict(X)

    results = {
        "speech_rate_score": y_pred[0][0],
        "energy_score": y_pred[0][1],
        "confidence_score": y_pred[0][2],
        "pitch_mean": y_pred[0][3],
        "pitch_std": y_pred[0][4],
    }

   
    os.remove(audio_path)

    return results
```
